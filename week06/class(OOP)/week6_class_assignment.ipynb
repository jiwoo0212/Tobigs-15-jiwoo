```python
# 유저 생성 클래스
class User():
    def __init__(self, user_name, stable_message, stable_music, friends_list, plus_friends_list):
        self.user_name = user_name
        self.stable_message = stable_message
        self.stable_music = stable_music
        self.friends_list = friends_list
        self.plus_friends_list = plus_friends_list

    def __str__(self):
        return '''My name is {}
        My stable message is {}
        My stable music is {}
        My friends list is {}
        My plus friends list is {}'''.format(self.user_name, self.stable_message, self.stable_music, self.friends_list,
                                             self.plus_friends_list)

    def get_user_name(self):
        return self.user_name

    def get_stable_message(self):
        return self.stable_message

    def get_stable_music(self):
        return self.stable_music

    def set_stable_message(self, stable_message):
        self.stable_message = stable_message

    def set_stable_music(self, stable_music):
        self.stable_music = stable_music

# 나의 친구목록에 존재하는 사람인지 구별하는 클래스
class Is_in():
    def __init__(self, user_a, user_b):
        self.user_a = user_a
        self.user_b = user_b

    def is_in_list(self, other):
        pass


class Is_in_friends(Is_in):
    def __init__(self, user_a, user_b):
        super().__init__(user_a, user_b)

    def is_in_list(self):
        if self.user_b.user_name in self.user_a.friends_list:
            return True
        else:
            return False


class Is_in_pfriends(Is_in):
    def __init__(self, user_a, user_b):
        super().__init__(user_a, user_b)

    def is_in_list(self):
        if self.user_b.user_name in self.user_a.plus_friends_list:
            return True
        else:
            return False


# 채팅하는 클래스
from abc import *


class Chat(metaclass=ABCMeta):

    @staticmethod
    def chat_with(self):
        pass


class Chat_friends(Chat):
    def __init__(self, user_a, user_b):
        self.user_a = user_a
        self.user_b = user_b

    def chat_with(self):
        if Is_in_friends(self.user_a, self.user_b).is_in_list():
            print('{} is talking with {}!!!'.format(self.user_a.user_name, self.user_b.user_name))
        else:
            print('Warning: Unknown User!!')
            yn = input('Keep this chat?(Y/N) :')
            if yn == 'Y':
                print('{} is talking with {}!!!'.format(self.user_a.user_name, self.user_b.user_name))
            else:
                print('Close this chat')


class Chat_plus_friends(Chat):
    def __init__(self, user_a, user_b):
        self.user_a = user_a
        self.user_b = user_b

    def chat_with(self):
        if Is_in_pfriends(self.user_a, self.user_b).is_in_list():
            print('{} is talking with {}!!!'.format(self.user_a.user_name, self.user_b.user_name))
        else:
            print('Warning: Unknown PLUS Friends!!')
            yn = input('Keep this chat?(Y/N) :')
            if yn == 'Y':
                print('{} is talking with {}!!!'.format(self.user_a.user_name, self.user_b.user_name))
            else:
                print('Close this chat')


a_friends = ['jiyoon', 'heeji', 'mincheol', 'minsu', 'suzy']
a_plus_friends = ['Tobigs', 'innisfree', 'Delta', 'Kakao Pay']

b_friends = ['jiwoo', 'heeji', 'mincheol', 'minsu']
b_plus_friends = ['Tobigs', 'Kakao friends']

c_friends = []
c_plus_friends = []

jiwoo = User('jiwoo', 'welcome 2021', 'Celebrity_IU', a_friends, a_plus_friends)
print(jiwoo, '\n')
suzy = User('suzy', 'music is my life....', 'Spaceman_NickJonas', b_friends, b_plus_friends)
print(suzy, '\n')

chungha = User('chungha', '', '', c_friends, c_plus_friends)
print(chungha, '\n')

chat1 = Chat_friends(jiwoo, suzy)
chat1.chat_with()

chat2 = Chat_friends(jiwoo, chungha)
chat2.chat_with()
```

위는 유저를 생성하는 클래스인 User, 해당 유저가 친구목록에 속하는 사람인지 구별하는 클래스인 Is_in, 유저끼리 채팅을 하는 클래스인 Chat 으로 이루어져 있습니다.

Is_in 이라는 추상클래스를 만들고, 파악하려는 대상이 유저인 경우 / 플러스 친구인 경우 두 개의 클래스가 상속받게합니다.
Chat이라는 추상 클래스를 만들고,  유저와 채팅하는 경우 / 플러스 친구와 채팅하는 두 개의 클래스가 상속받게합니다. 

SOLID 원칙은 아래와 같습니다. 
- Single Responsiblity Principle (단일 책임 원칙) : 소프트웨어의 설계 부품(클래스, 함수 등)은 단 하나의 책임만을 가져야 한다.
- Open-Closed Principle (개방-패쇄 원칙) : 기존의 코드를 변경하지 않고(Closed) 기능을 수정하거나 추가할 수 있도록(Open) 설계해야 한다.
- LSP(Liskov Substitusion Principle) 리스코프 치환 법칙 : 자식 클래스는 언제나 자신의 부모클래스와 교체할 수 있다는 원칙
- ISP(Interface Segregation Principle) 인터페이스 분리 원칙 : 클래스에서 사용하지 않는(상관없는) 메서드는 분리해야 한다.
- DIP(Dependency Inversion Principle) 의존성 역전 법칙 : 부모 클래스는 자식 클래스의 구현에 의존해서는 안된다, 자식 클래스 코드 변경 또는 자식 클래스 변경시, 부모 클래스 코드를 변경해야 하는 상황을 만들면 안된다.

Is_in, Chat 두 개의 추상 클래스를 구성해서 일반 유저와 플러스 친구인 경우를 분리했다는 점, Is_in 클래스를 만들어서 추후 변경 사항이 있으면 is_in 클래스에서 수정하면 된다는 점, 모든 상위 클래스는 하위 클래스에 의존되지 않도록 구성했다는 점에서 SOLID 원칙을 만족한다고 볼 수 있습니다.
